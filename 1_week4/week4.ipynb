```python
import numpy as np
from numpy.random import gamma
import matplotlib.pyplot as plt
```

Генерируем выборку из гамма-распределения определенного объема. Фпв имеет вид $$f(x;k,\theta)=x^{k-1}\frac{e^{-x/\theta}}{\theta^k\Gamma(k)}$$
Отсюда: $$E[X]=k\theta$$ $$D[X]=k\theta^2$$

```python
k = 1
t = 10
nu = k*t #Это для нормального распределения
sig2 = k*t**2
r = gamma(3,10,size=10000)
plt.hist(r,20);
```

Далее сгененируем по 1000 выборок объемов 5,10,50. Т.е. нужно получить три массива 5х1000, 10х1000, 50х1000. Ну и сразу считаем выборочное среднее, чтобы далеко не ходить


```python
X5  = np.array([ (gamma(k,t, size=5 )).mean() for i in range(1000)])
X10 = np.array([ (gamma(k,t, size=10)).mean() for i in range(1000)])
X50 = np.array([ (gamma(k,t, size=50)).mean() for i in range(1000)])
```

```python
a1 = plt.hist(X5,density=True)
sigt = sig2/5
N5 = 1/sigt/np.sqrt(2*np.pi) * np.exp(-(a1[1]-nu)**2/2/sigt)
plt.plot(a1[1],N5)
```

```python
a1 = plt.hist(X10,density=True);
sigt = sig2/10
N10 = 1/sigt/np.sqrt(2*np.pi) * np.exp(-(a1[1]-nu)**2/2/sigt)
plt.plot(a1[1],N10)
```

```python
plt.hist(X50,density=True);
sigt = sig2/50
N50 = 1/sigt/np.sqrt(2*np.pi) * np.exp(-(a1[1]-nu)**2/2/sigt)
plt.plot(a1[1],N50)
```

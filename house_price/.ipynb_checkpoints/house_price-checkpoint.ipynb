```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_squared_log_error as msel 
from sklearn.metrics import r2_score
import seaborn as sns
import matplotlib.pyplot as plt
```

<h2>Exploratory data analysis</h2>

**Look and insert missing values**

```python
data = pd.read_csv('train.csv')
y = data.iloc[:,80]
X = data.drop(columns=['Id','SalePrice'])

def drop_data(X):    

    missing = X.isnull().sum()
    missing = missing[missing >0]
    missing.sort_values(inplace=True)
    
    #We will drop all variables with missing values greater than 0.5
    tmp = missing[missing/X.shape[0] > .5].index
    X.drop(tmp, axis=1, inplace=True)
    
    return X,missing
    
X,missing = drop_data(X)     
```

```python
missing.plot.bar()
print('procentage of missing values: \n\n',missing/X.shape[0]*100)
```

Dividing data into categorical and numerical

```python
def divid_ncat(X):
    dictX = dict(X.dtypes)

    cate_keys = []
    numeric_key = []
    for keyX in dictX.keys():
        if (str(dictX[keyX]) != 'object'):
            numeric_key.append(keyX)
        elif str(dictX[keyX]) == 'object':
            cate_keys.append(keyX)
    return cate_keys,numeric_key

def replace_missing(Xnum):
    cols_with_missing = [col for col in Xnum.columns if Xnum[col].isna().any()] #colums with missings
    for col in cols_with_missing:
        Xnum[col] = Xnum[col].fillna(0) #Replace by nulls
    return Xnum

cate_keys,numeric_key = divid_ncat(X)

Xnum  = X[numeric_key]
Xcate = X[cate_keys]

Xnum = replace_missing(Xnum)
```

```python
corr = Xnum.corr()
plt.subplots(figsize=(15,12))
sns.heatmap(corr,vmax=.9,cmap='Blues',square=True)
Xnum.corrwith(y).sort_values(ascending=False)
```

Now we divide data to train and test 

```python
Xnum_train,Xnum_test,y_train,y_test = train_test_split(Xnum.to_numpy(),y.to_numpy(),test_size=.25)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(Xnum_train,y_train)

y_pred_train = np.int64( np.abs( reg.predict(Xnum_train) ))
y_pred_test = np.int64( np.abs( reg.predict(Xnum_test) ) ) #One value in regression prediction is negative
```

```python
def metrics_score(y_pred,y,strng):
    print(np.int64(y[:15]/1000))
    print(np.int64(y_pred[:15]/1000))    
    
    print(strng)
    print('rmse = {:.2f}'.format(mse(y,y_pred,squared=False))  )
    print('R2 = {:.2f}'.format(r2_score(y,y_pred))  )
    print('logarifmic rmse = {:.2f}'.format(np.sqrt( msel(y,y_pred) ) ),'\n'  )
    
```

```python
metrics_score(y_pred_train,y_train,'errors for train data:')
metrics_score(y_pred_test,y_test,'errors for TEST data:')
```

<h3>This was be a simple linear regression without parameters finding. At now add cross validation and parameters search.</h3>

```python
from sklearn.linear_model import SGDRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import metrics


regSGD = SGDRegressor()

parameters_grid = {
    'alpha': np.logspace(-1,5,num=15),
    'fit_intercept': [False],
    'penalty': ['l1','l2','elasticnet'],
    'shuffle': [False],
    'random_state':[12]
}
```

```python
def mean_squared_log_ABS_error(y_true,y_pred):
    y_pred = np.abs(y_pred)
    return metrics.mean_squared_log_error(y_true,y_pred,squared=True)
    
ABS_scorer = metrics.make_scorer(
    mean_squared_log_ABS_error,greater_is_better=False)

grid_cv = GridSearchCV(regSGD,parameters_grid,scoring='neg_mean_squared_log_error')
```

```python
grid_cv.fit(np.hstack((Xnum.to_numpy(),Xcate_trf)),y.to_numpy())
```

<h3>Predict on test data</h3>

```python
def predictOnTestData(reg,scale_transform=True,log_transform=False):

    data = pd.read_csv('test.csv')
    Id = data['Id']
    X2 = data.drop(columns=['Id'])
    X2,missing2 = drop_data(X2) 

    cate_keys,numeric_key = divid_ncat(X2)
    Xnum2  = X2[numeric_key]
    Xcate2 = X2[cate_keys]

    Xnum2 = replace_missing(Xnum2)

    if scale_transform:
        scale_transform = StandardScaler()
        Xnum2 = scale_transform.fit_transform(Xnum2)

    X2cate_trf = encoder.transform(Xcate2.T.to_dict().values())

    Xtest_final = np.hstack((Xnum2,X2cate_trf))
    
    y_pred_test = np.abs( reg.predict(Xtest_final) )
    
    if log_transform:
        y_pred_test = np.exp(y_pred_test)

    d = {'Id':Id,'SalePrice':y_pred_test}
    df = pd.DataFrame(data=d)
    df.to_csv('y_pred_test.csv',index=False)
    df.head()
    df.shape
    
predictOnTestData(reg)
```

```python
#As we can accidentally change something, reobtain train data
cate_keys,numeric_key = divid_ncat(X)

Xnum  = X[numeric_key]
Xcate = X[cate_keys]

Xnum = replace_missing(Xnum)

#For random forest it's no necessary to scale numeric variables

for i in cate_keys:
    Xcate.loc[:,i] = Xcate.loc[:,i].astype(str)
encoder = DictVectorizer(sparse=False)
Xcate_trf = encoder.fit_transform(Xcate.T.to_dict().values())

X_concat = np.hstack((Xnum.to_numpy(),Xcate_trf))
y_concat = y.to_numpy()
X_train,X_test,y_train,y_test = train_test_split(X_concat,y_concat,test_size=.25)
```
